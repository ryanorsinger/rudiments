```python
import pandas as pd
import seaborn as sns
from pydataset import data
```

```python
df = data("tips")
df.head()
```

```python
df.total_bill.hist()
```

```python
def bin_bill(bill):
    if bill <= 10:
        return "Under10"
    elif bill <= 20:
        return "Above10Under20"
    elif bill <= 30:
        return "Above20Under30"
    else:
        return "Big Spender"
            

df["binned_bill"] = df.total_bill.apply(bin_bill)
```

```python
df.head()
```

```python
# Scatterplots ain't so easy to read w/ a category as the x or y
# We don't want to use scatterplots with categories, unless the category is the hue argument or col argument
sns.relplot(x="binned_bill", y="tip", data=df)
```

```python
# We absolutely love scatterplots for continuous x and continuous y
sns.relplot(x="total_bill", y="tip", data=df)
```

```python
# We absolutely love scatterplots for continuous x and continuous y
sns.relplot(x="total_bill", y="tip", data=df, hue="smoker")
```

```python
# If our X is categorical or a bin, a boxplot tells us a LOT about the y for each category X
sns.boxplot(x="binned_bill", y="tip", data=df)
```

```python
sns.violinplot(x="binned_bill", y="tip", data=df)
```
