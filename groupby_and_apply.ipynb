```python
from pydataset import data
import pandas as pd
from scipy import stats
```

```python
df = data("tips")
df.head()
```

```python
def corr_tenure_and_total(df):
    return pd.Series(stats.pearsonr(df.tip, df.total_bill), index=['r', 'p'])
```

```python
df.groupby(["sex", "smoker", "time"]).apply(corr_tenure_and_total)
```

```python
df.groupby(["sex", "smoker", "time"]).apply(corr_tenure_and_total).reset_index()
```
